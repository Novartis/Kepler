```python
import Kepler.plotting as kplot
import Kepler.preprocessing as kpre

import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
integrated_path = "/da/dmp/cb/tysinem1/kidney_dataset_original/kidney_stx/integrated_spotlight_07212022.h5ad"
adata=sc.read_h5ad(integrated_path)
adata.obsm['spatial_fr'] = adata.obsm['spatial']
adata.obsm['spatial'] = adata.obsm['spatial_original']
```

```python
kplot.scspatial
```

```python
spatial_folder = "/da/dmp/cb/wuji3k/projects/dmpa-wuji3k-spatial-transcriptomics"
infoTable = pd.read_csv(f'{spatial_folder}/infotable_pilot.csv', sep=',')
infoTable
```

```python
infoTable_frozen_only = infoTable[infoTable['sample_type'] == 'frozen']
adata_res, gdf_dict = kpre.load_geojson_annotations(infoTable_frozen_only, adata, feature_list = ['Cyst', 'Dilated tubules'])
```

```python
adata_res.var_names = adata_res.var['symbol']
symbol_to_ensmusg=adata_res.var['gene_ids'].to_dict()
#adata_res.var_names = adata_res.var['gene_ids']
#adata_res.var
```

```python
library_id = infoTable['sample_names'][2]
gdf_dict[library_id]['gdf_sp'] = kpre.fill_gdf_classification(gdf_dict[library_id]['gdf_sp'])
gdf_colors = {'Cyst': 'red', 'Dilated tubules': 'yellow'}
```

```python
# gdf_dict['WD-07-CP86']
```

```python
fig,axs = plt.subplots(1,1,figsize=(15,10))
fig=kplot.scspatial(adata_res[adata_res.obs['sample_names'] == library_id], color=symbol_to_ensmusg['Vim'],title='Vim',library_id=library_id, ax=axs, show=True, gdf=gdf_dict[library_id]['gdf_sp'], gdf_colors=gdf_colors, cmap='jet', gdf_alpha=0.5)

plt.show()
```

```python
adata_res.obs.loc[(adata_res.obs['sample_names'] == library_id) & (adata_res.obs['distance_to_any_cyst_um'] < 550) & (adata_res.obs['distance_to_any_cyst_um'] > 100), 'cyst_bins'] = 'non-boundary'
adata_res.obs.loc[(adata_res.obs['distance_to_any_cyst_um'] < 100), 'cyst_bins'] = 'boundary'
```

```python
fig,axs = plt.subplots(1,1,figsize=(15,10))
fig=kplot.scspatial(adata_res[adata_res.obs['sample_names'] == library_id], color='cyst_bins',title='Vim',library_id=library_id, ax=axs, show=True, gdf=gdf_dict[library_id]['gdf_sp'], gdf_colors=gdf_colors, cmap='jet', gdf_alpha=0.5)

plt.show()
```
